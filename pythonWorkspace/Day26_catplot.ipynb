```python
import seaborn as sns
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib
```

```python
pd_tips = sns.load_dataset('tips')
pd_tips.head()
```

```python
sns.catplot(data=pd_tips, x='day', y='total_bill', hue='smoker',dodge=True, kind='swarm') #jitter 퍼짐정도 조절,dodge: 따로 빼줌
# sns.relplot(data=pd_tips, x='day', y='total_bill')
sns.catplot(data=pd_tips, x='total_bill', y='day', hue='smoker',dodge=True, kind='swarm')
```
