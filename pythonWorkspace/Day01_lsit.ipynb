```python
lista = [i for i in range(10)]
listb = [i for i in range(100,110)]
print(lista)
print(listb)

#in    list안에 값이 있는지 확인하는 
if 5 in lista:
    print('True')
else:
    print('False')
#not in 안에 값이 없는지 확인

#concatenation 리스트 붙이기
print(lista +listb)

#*  리스트 전체 반복
lista * 2 
```

```python
listd = list('abcdefghicjkmnlopqrstuvw')
print(listd)
print(listd[3])# 인덱스 추출
print(listd[0:6])#슬라이싱
print(listd[0:6:2]) # 스탭
print(len(listd))
print(min(listd))
print(max(listd))
print(listd.index('c',3))
#print(listd.index('c',3,6))
#listd[2] = 'Z'  값 update
#print(listd[2])
print('count: ', listd.count('c'))
```

```python
s = 'While The Python Language Reference describes the exact syntax and semantics of the Python language, this library reference manual describes the standard library that is distributed with Python. It also describes some of the optional components that are commonly included in Python distributions.'
print(s.lower()) # 소문자로 변환
#print(s)
list_a = list(s)

#print(list_a)

set_a = set(list_a)
print(set_a)

exclude_list = ['.',' ']

for i in set_a:
    if i not in exclude_list:
        print(i,list_a.count(i))
```

```python
list_a = [1,2,3]
list_b = list('abc')
d1 = {'a' : 1, 'b' : 2, 'c' : 3}
print(list_a)
print(list_b)
print(d1)

d2 = dict(zip(list_b, list_a)) # (value , key) 로 리스트를 딕셔너리로 묶어줌
print(d2)
print(d2.items())

d3 = dict([('a', 1), ('b', 2), ('c', 3)])
print(d3)

print(d3['c']) # key 값으로 value 값을 조회한다.
print(d3.get('c')) # 위와 같이 key 값으로 value값을 조회하는ㄷ데 없는 키값을 넣어도 에러가 나지 않는다
d3['y'] = 15 #딕셔너리 값 추가
print(d3)
```

```python
for i in d3:
    print(i, d3[i])
    
print(d3.keys())
print(d3.values())

for i in d3.values():
    print(i)

print(d3)

d3['y'] = 20
print(d3)
```
