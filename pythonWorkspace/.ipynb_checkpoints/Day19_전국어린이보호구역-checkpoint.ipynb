```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

pd_rawdata = pd.read_csv('전국어린이보호구역표준데이터.csv', encoding = 'cp949')
# print(pd_data.shape)
# print(pd_data.dtypes)
# pd_data.head(5)
colums_selection = ['시설종류','관할경찰서명','CCTV설치여부','CCTV설치대수','제공기관명','보호구역도로폭']
pd_data = pd_rawdata[colums_selection]
del pd_rawdata
```

```python
#EDA : Exploratory Data Analysis

print(pd_data.columns)
print(pd_data.dtypes)

print(pd_data['보호구역도로폭'].value_counts())
print(pd_data['보호구역도로폭'].isna().value_counts())
pd_data['시설종류'] = pd_data['시설종류'].astype('category')
pd_data['CCTV설치여부'] = pd_data['CCTV설치여부'].astype('category')
# print(pd_data.dtypes)
pd_data.describe(include = 'all')
```

```python
#결측지 확인 

# print(pd_data.shape)
# print(pd_data.describe(include='all'))

for e in pd_data:
    print(e,pd_data[e].hasnans)
    if pd_data[e].hasnans:
        print(pd_data[e].isna().value_counts())
```

```python
#CCTV 설치대수 : 결측치 처리
#CCTV 설치여부 Y --> 1
#CCTV 설치여부 N --> 0
print(pd_data['CCTV설치대수'].isna().value_counts())
filter_y = pd_data['CCTV설치여부'] == 'Y'
filter_n = pd_data['CCTV설치여부'] == 'N'

pd_data

# pd_data.replace({'CCTV설치여부':'Y','CCTV설치대수': np.NaN},1, inplace=True)
# pd_data.replace({'CCTV설치여부':'N','CCTV설치대수': np.NaN},0, inplace=True)

print(pd_data['CCTV설치대수'].isna().value_counts())
```

```python
def my_fun1(x):
    if type(x) == type(' '):
        if'~' in x:
            m = np.array(x.split('~')).astype(np.float64).mean()
            return str(m)
    
#도로폭 : 오류치 처리 (숫자~숫자 형태의 문자열 --> 평균치 np.float64 값으로 대체)
# print(pd_data['보호구역도로폭'].value_counts(my_fun1))
y = pd_data['보호구역도로폭'].apply(my_fun1)
# print(y.dtype)
# print(y.value_counts())


#도로폭 : 결측지 처리
#전체 도로폭 평균치로 대체.
print(pd_data['보호구역도로폭'].isna().value_counts())
y = y.astype(np.float64)
y = y.replace(np.NaN, y.mean())
print(y.isna().value_counts())
pd_data['보호구역도로폭'] = y
```

### 시설종류 
* dtype : category
* value:
* 걸측지 없음
### 관할경찰서명
* dtype : object
* 결측지 없음
### CCTV 설치여부
* dtype : category
* value : Y , N
* 결측지 없음
### CCTV 설치대수
* dtype : np.int64
* min / max : 0 / _
* 결측지 (개, 처리 후 결측지 없음)
    cctv 설치여부 Y -> 1
    cctv 설치여부 N -> 0    
### 보호구역 도로폭
* dtype : np.float64
* min / max
* 오류치 (     개 , 처리후 오류치 없음)
    도로폭 : 오류치 처리 (숫자~숫자 형태의 문자열 --> 평균치 np.float64 값으로 대체)
* 결측지 (개 처리후 결측지 없음
    전체 평균값으로 대체

- 시설종류별 cctv 설치 여부 카운트 / 대수
- 광역시도별, 시설종류별 카운트
- 관할 경찰서별 cctv 대수
- cctv평균 설치 대수
- 도로폭 / cctv 설치대수 상관도
