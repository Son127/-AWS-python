```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib
```

```python
xs = np.linspace(0, 50 ,100)
ys = np.sin(xs)

#figure 
figure,axe = plt.subplots()
figure.set_facecolor('blue') # 
figure.suptitle('Figure1') #
figure.supxlabel('X') # x  
figure.supylabel('Y')
figure.text(0.5,0.7,'test')
figure.set_size_inches(20, 5)

#axe 
axe.plot(xs,ys)
axe.set_title('my axe title')
axe.set_xlabel('axe_x')
axe.set_ylim(0,1) #   (? ~ ?)
axe.set_xlim(0,30)
axe.set_xticks(np.linspace(0,30,20))
axe.set_xticklabels(list('abcd'))

print(type(figure))
print(type(axe))
```
