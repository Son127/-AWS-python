```python
import seaborn as sns
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib
import requests

current_font_list = matplotlib.rcParams['font.family']
font_path = "C:\Windows\Fonts\gulim.ttc"
kfont = matplotlib.font_manager.FontProperties(fname=font_path).get_name()

matplotlib.rcParams['font.family'] = [kfont] + current_font_list

abalone_dt = pd.read_csv('https://archive.ics.uci.edu/ml/machine-learning-databases/abalone/abalone.data')

abalone_dt.columns = ['Sex', 'Length', 'Diameter', 'Height','Whole weight','Shucked weight','Viscera weight'
                           ,'Shell weight','Rings']
abalone_df = abalone_dt[abalone_dt['Sex'] != 'I']
abalone_df

weight_data = abalone_df.groupby('Sex')[['Whole weight', 'Shucked weight', 'Viscera weight', 'Shell weight']].mean()
print(weight_data)
```

```python
fig, axes = plt.subplots(nrows=2, ncols=3, figsize=(14, 10))

sns.barplot(x=weight_data.index, y='Whole weight', data=weight_data, ax=axes[0,0])
axes[0, 0].set_title('암,수 전체 무게 비교')

sns.barplot(x=weight_data.index, y='Shucked weight', data=weight_data, ax=axes[0, 1])
axes[0,1].set_title('암,수 살점 무게 비교')

sns.barplot(x=weight_data.index, y='Viscera weight', data=weight_data, ax=axes[0, 2])
axes[0,2].set_title('암수 내장 무게 비교')

sns.barplot(x=weight_data.index, y='Shell weight', data=weight_data, ax=axes[1, 0])
axes[1, 0].set_title('암,수  껍질 무게 비교')

sns.relplot

```
