```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
xs = np.linspace(0, 50 ,100)
ys = np.sin(xs)

#figure 설정
figure,axe = plt.subplots()
figure.set_facecolor('blue') #배경 색
figure.suptitle('Figure1') #타이틀
figure.supxlabel('X') # x 축 타이틀
figure.supylabel('Y')
figure.text(0.5,0.7,'test')
figure.set_size_inches(20, 5)

#axe 설정
axe.plot(xs,ys)
axe.set_title('my axe title')
axe.set_xlabel('axe_x')
axe.set_ylim(0,1) # 확대하는 기능 (? ~ ?)
axe.set_xlim(0,30)
axe.set_xticks(np.linspace(0,30,20))
axe.set_xticklabels(list('abcd'))

print(type(figure))
print(type(axe))
```
